```python
from pathlib import Path
import shutil

# Define paths (same logic as config.py but inline for convenience)
BASE_DIR = Path("/home/jovyan/work")  # root of your repo inside the Jupyter container

INPUT_DIR = BASE_DIR / "data" / "input"
DONE_DIR = BASE_DIR / "data" / "done"
ERROR_DIR = BASE_DIR / "data" / "error"
OUTPUT_DIR = BASE_DIR / "output" / "daily_summary"

# --- 1Ô∏è‚É£ Clear output daily_summary ---
if OUTPUT_DIR.exists():
    for file in OUTPUT_DIR.glob("*.csv"):
        file.unlink()
    print("‚úî Output cleared:", OUTPUT_DIR)
else:
    print("‚ö† Output directory does not exist:", OUTPUT_DIR)

# --- 2Ô∏è‚É£ Move files from done/ ‚Üí input/ ---
if DONE_DIR.exists():
    for file in DONE_DIR.glob("*"):
        shutil.move(str(file), INPUT_DIR / file.name)
    print("‚úî Done ‚Üí Input")
else:
    print("‚ö† Done directory does not exist:", DONE_DIR)

# --- 3Ô∏è‚É£ Move files from error/ ‚Üí input/ ---
if ERROR_DIR.exists():
    for file in ERROR_DIR.glob("*"):
        shutil.move(str(file), INPUT_DIR / file.name)
    print("‚úî Error ‚Üí Input")
else:
    print("‚ö† Error directory does not exist:", ERROR_DIR)

print("üéâ Reset completed!")
```
